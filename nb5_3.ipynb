```python
#1. 연결(+) 연산

a = [1, 2, 3]
b = [4, 5]
result = a + b
print(result) # [1, 2, 3, 4, 5]
print(a)
print(b)
```

```python
#2. 반복(*) 연산

nums = [10, 20]
result = nums * 3
print(result) # [10, 20, 10, 20, 10, 20]

# [30, 60] Numpy 라이브러리로 행렬 객체를 만들어 사용
```

```python
#3. in, not in 연산

fruits = ["apple", "banana", "orange"]
print("banana" in fruits)

print("grape" in fruits)
print("grape" not in fruits)
```

```python
#4. 인덱싱

nums = [10, 20, 30, 40]

print(nums[0])
print(nums[-1]) # -1은 맨 끝의 값

# 인덱스 할당 연산
nums[2] = 99 # 리스트는 가변 객체
print(nums)

n2 = (3,5)
print(n2[0])
print(n2[-1])
n2[0] = 999 #인덱스 할당 지원하지 않아서 에러 발생
#tuple = 리스트의 읽기 전용 타입/리스트의 불변 타입
```

```python
#5. 슬라이싱 / 파이썬에서 [시작:끝]을 적을 때 '끝'에 적는 숫자는 포함하지 않고, 그 바로 앞까지만 잘라라!라는 절대적인 규칙이 있음

nums = [10, 20, 30, 40, 50]

print(nums[1:4]) # 2번째부터 4번째까지 표기 [20, 30, 40]
print(nums[:3]) # 처음부터 3번째까지 표기 [10, 20, 30] 
print(nums[2:]) # 3번째부터 끝까지 표기 [30, 40, 50]
print(nums) #  [10, 20, 30, 40, 50] 슬라이싱 연산 후에도 리스트 자신은 변경되지 않는다.

#data[:]      # 전체
#data[:5]     # 앞 5개
#data[5:]     # 5번째 이후
#data[-5:]    # 마지막 5개
#data[::-1]   # 역순
```

```python
#6. 리스트 비교연산

a = [1, 2, 3]
b = [1, 2, 3]
c = [1, 2, 4]

print(a == b) # True
print(a == c) # False
```

```python
#7. 리스트를 인자값으로 자주 사용되는 내장함수 / import 불필요

nums = [10, 20, 30, 40]

print(len(nums))
nums = [10, 20, 30, 40]

print(max(nums))
print(min(nums))
print(sum(nums))
```

```python
#8. 리스트 메서드

# append() : 맨 뒤에 1개 추가
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)


# insert() : 원하는 위치에 추가
fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)


# extend() : 여러 개 추가
nums = [1, 2, 3]
nums.extend([4, 5, 6])
print(nums)


# remove() : 값으로 삭제
fruits = ["apple", "banana", "orange"]
fruits.remove("banana")
print(fruits)
# 수정은 인덱스 할당, 슬라이싱 할당으로 수정


# pop() : 인덱스로 삭제 후 반환
nums = [10, 20, 30]
result = nums.pop() # 마지막 인덱스 삭제 후 반환
print(result)
print(nums)


nums = [10, 20, 30]
result = nums.pop(1) # 인덱스1 삭제 후 반환
print(result)
print(nums)


# clear() : 전체 삭제
nums = [10, 20, 30]
nums.clear()
print(nums)


# del 문: 삭제
nums = [10, 20, 30, 40]
del nums[1]
print(nums)


# reverse() : 순서 뒤집기, 정렬과 상관없이 현재 순서를 거꾸로 만든다.
nums = [10, 20, 30, 40]
nums.reverse()
print(nums)


# count() : 특정 값이 몇개인지
nums = [10, 20, 10, 30, 10]
print(nums.count(10)) # 3


# index() : 특정 값의 인덱스
fruits = ["apple", "banana", "orange"]
print(fruits.index("banana")) # 1


# 정렬 메서드
nums = [50, 10, 40, 20]
nums.sort() # 오름차순
print(nums)

nums = [50, 10, 40, 20]
nums.sort(reverse=True) # 내림차순
print(nums)
```

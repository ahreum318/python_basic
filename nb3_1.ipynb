```python
# 1. 정수 (int)
a = 10
print(a)
print(type(a))
```

```python
# 2. 부동 소수점 수 (float) : 오차를 가진다
b = 3.14
print(b)
print(type(b))

print(0.1 + 0.2) #0.100...n + 0.200...n : 16번째 자리까지는 정확도를 가지나 이후는 근사치 표시
```

```python
# 3. 복소수 (complex)
c = 3 + 4j
print(c)
print(type(c))
```

```python
# 4. 논리값 (bool)
d = True
print(d)
print(type(d))

print(3>5) #False
```

```python
# 5. 문자열 (str)
e = "Python"
print(e)
print(type(e))

print(e[1]) #1번 글자를 출력해라 (0번부터 시작)

e[1] = "K" #"Python"은 문자열객체 == 불변객체, 바꿀 수 없으므로 에러남
```

```python
# 6. 리스트 (list) / 객체 여러가지를 가질 수 있음
f = [1, 2, 3]
print(f)
print(type(f))

print(f[1])

f[1] = "abc" #가변객체로 바꿀 수 있음
print(f)
```

```python
# 7. 튜플 (tuple) / 읽기 전용 list 타입이라고 생각
g = (1, 2, 3)
print(g)
print(type(g))

print(g[1])

g[1] = 9 #불변객체로 에러남
```

```python
# 8. 딕셔너리 (dict)
h = {"id": "admin", "pw": "1234"} #키가 복잡해질 수 있음 : (1,3,5,6): "1234"의 형태로 튜플 삽입 가능
print(h)
print(type(h))

print(h["id"])
print(h["pw"])

h["pw"] = "0000" #가변객체로 변경 가능, pw: 1234 -> 0000
print(h)
```

```python
# 9. 집합 (set)
i = {1, 2, 3, 2} #중복값 제외
print(i)
print(type(i))

print(i[0]) #집합은 전체만 불러올 수 있음
```

```python
# 10. 범위 (range)
j = range(0, 5)
print(j)
print(type(j))

print(list(j))

#index
```

```python
# 11. 바이트 (bytes)
k = b"abc"
print(k)
print(type(k))
```

```python
# 12. 바이트배열 (bytearray)
l = bytearray(b"abc")
print(l)
print(type(l))
```

```python
# 13. 없음 (NoneType)
m = None
print(m)
print(type(m))
```

```python
# 14. 함수 (function)
def hello():
    print("Hello")
    # 탭: 위 식에 소속되어 있는 코드

def hello():
    print("Hello")
    print("bye")
print("hi")

print(hello) #함수 안에 저장되는 것은 코드, 함수 코드는 호출 가능함
hello() #이렇게 해야 함수 내용이 출력됨
print(type(hello))

a = 10
a()
# 에러남 10은 기능이 없는 정수(int)기 때문에
```

```python
# 15. 모듈 (module)
import math

print(math)
print(type(math))
```

```python
# 16. 클래스 (type)
class Person:
    pass

print(Person)
print(type(Person))
```

```python
#for문 + in 연산
# → 쓰는 이유: 알맹이만 건지려고, 알맹이 건진 후에 가공하려고

fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)
    print("hello")


for fruit in fruits:
    print(fruit.upper())

for fruit in fruits:
    if len(fruit) >= 6:
        print(f"길다! : {fruit}")
```

```python
colors = ("red", "green", "blue")

for color in colors:
    print(color)
```

```python
nums = {10, 20, 30}

for num in nums:
    print(num)
```

```python
student = {
    "name": "홍길동",
    "age": 20
}


# 키
for key in student:
    print(key)
    
# 값
for value in student.values():
    print(value)

# 키 + 값 → 언패킹 발생
for key, value in student.items():
    print(key, value)
```

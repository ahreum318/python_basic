```python
# 이스케이프 문자

# \n: 줄바꿈
print("Hello\nPython")

# \t: 탭 이동
print("이름\t나이")
print("김아름\t26")

# \\: 역슬래시 그대로 출력하고 싶을 때
print("파일 경로: C:\\Users\\Desktop")
```

```python
# Raw 문자열 : 문자열 안의 이스케이프 문자(\n, \t, \\ 등)를 특별하게 해석하지 않고 그대로 표현
# 문자열의 시작 따옴표 앞에 r붙임

print(r"C:\Users\data")

# 해결은 되나 번거로움
path = "C:\\users\\name\\desktop"
print(path)
```

```python
# f-string : 문자열 안에 변수 표현식{변수} 사용
name = "Tom"
print(f"이름은 {name}이다")
print("이름은 "+name+"이다")

age = 20
print("나이는 " + str(age)+"입니다")
print(f"나이는 {age}이다")
```

```python
#대소문자 변환
text = "Python Programming"

print(text.upper())      # 모두 대문자
print(text.lower())      # 모두 소문자
print(text.capitalize()) # 첫 글자만 대문자
print(text.title())      # 각 단어 첫 글자 대문자
```

```python
#공백제거
text = "   Pyth on   "

print(text.strip())   # 양쪽 공백 제거
print(text.lstrip())  # 왼쪽 공백 제거
print(text.rstrip())  # 오른쪽 공백 제거
#글자 사이 공백 있을 시 무조건 유지됨
```

```python
#문자열 분리

data = "apple,banana,orange"
result = data.split(",") #콤마를 기준으로 문자열을 나눠줘
print(result) # ['apple', 'banana', 'orange']
```

```python
#문자열 결합

fruits = ["apple", "banana", "orange"]
result = ", ".join(fruits)
print(result)

student = ["방주성", "박경필", "김주영"]
result = "님, ".join(student)
print(result) #join은 단어 사이를 연결하는 메서드이기 때문에 마지막 사람은 '님'이 붙지 않음

result = "님, ".join(student) + "님"
print(result)
```

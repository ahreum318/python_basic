```python
#딕셔너리 

# - key : value 구조
# - 순서 O (Python 3.7 이상)
# - key는 중복 불가, value는 중복 가능
```

```python
# 빈 딕셔너리
d1 = {}
d2 = dict()

print(d1)
print(d2)
```

```python
#1. dict 자료형 생성, 다루기

#방법1: {} 사용
d1 = {"a": 1, "b": 2}
print(d1)

student = {
    "name": "홍길동",
    "age": 20,
    "score": 90
}

print(student)
print(type(student))
```

```python
#방법2: dict() 사용
d2 = dict(a=1, b=2)
print(d2)

person = dict(name="김철수", age=30, city="서울")

print(person)
print(type(person))
```

```python
#방법3: [()]를 dict으로 변환
d3 = dict([("a", 1), ("b", 2)])
print(d3)

data = [
    ("apple", 1000),
    ("banana", 2000),
    ("orange", 3000)
]

fruit = dict(data)

print(fruit)
print(type(fruit))
```

```python
# dict 의 키 값은 언제나 불변이어야 함

d = {"name": "홍길동"}
print(d)

d = {100: "학생"}
print(d)

d = {
    (1, 2): "좌표"
}
print(d)

d = {
    [1, 2]: "리스트"
} # error
print(d)
```

```python
# 연산

student = {"name": "홍길동"}
print(student)

# 요소조회
print(student["name"])

# 추가
student["age"] = 20
print(student)

# 수정
student["age"] = 30
print(student)

# 삭제
del student["age"]
print(student)

# in 연산
print("name" in student)
print("age" in student)
```

```python
# 메서드

student = {"name": "홍길동"}

print(student.get("name"))
print(student.get("age")) # 없으면 None 반환

student = {"name": "홍길동", "age": 20}
print(student.keys())
print(student.values())
print(student.items())

student.update({"age": 30, "score": 90})
print(student)

age = student.pop("age")
print(age)
print(student)

student.clear()
print(student)
```

```python
#함수

student = {
    "name": "홍길동",
    "age": 20,
    "score": 90
}
print(len(student))

s = {
    "c": 3,
    "a": 1,
    "b": 2
}

print(sorted(s)) # 반환값은 정렬된 키의 리스트
print(s) # 원본 변화 없음

print(list(s)) # 반환값은 키의 리스트
```

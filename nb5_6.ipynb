```python
#딕셔너리 

# - key : value 구조
# - 순서 O (Python 3.7 이상)
# - key는 중복 불가, value는 중복 가능
```

```python
#1. dict 자료형 생성, 다루기

#방법1
d1 = {"a": 1, "b": 2}
print(d1)

#방법2
d2 = dict(a=1, b=2)
print(d2)

#방법3
d3 = dict([("a", 1), ("b", 2)])
print(d3)
```

```python
# dict 의 키 값은 언제나 불변이어야 함

d1 = {(1, 2): "루피"}
d1[(1, 2)]
```

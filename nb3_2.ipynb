```python
#1. 산술연산자

a = 5
b = 2

print(a + b)   # 7
print(a // b)  # 2 (몫)
print(a % b)   # 1 (나머지)
print(5 ** 2)  # 25 (제곱)
```

```python
#2. 비교연산자

print(5 > 3)      # True
print(5 != 3)     # True
print(5 == 3)     # False
print(5 <= 3)     # False
```

```python
#3. 논리연산자

print(True and False)   # False
print(True or False)    # True
print(not True)         # False

print(10>5 and 10 == 7) # True and False
```

```python
#4. 멤버십연산자

print(3 in [1, 2, 3])       # 리스트 객체 안에 3이라는 값이 있느냐 # True 
a = [1, 2, 3]
print(4 in a)  # False
print(4 not in a)   # True
```

```python
#5. 식별연산자

a = [1, 2]
b = a # b는 [1, 2]를 가리킴

a is b      # True
a == b      # True

c = [1, 2]
d = [1, 2]

c is d      # False
c == d      # True

# == : 객체의 값만 비교 
# is : 객체 비교
```

```python
#6. 삼항연산자

score = 50
result = "합격" if score >= 60 else "불합격"
print(result)
```

```python
#7. 대입문

a = 10

# a = a+ 5와 같은 표현(편한거 사용하면 된다) 
a += 5 
print(a)  # 15
```

```python
#8-1. 월러스 연산자

n = 10
result = (x = n>5)   # 대입하면서 값 반환
print(result)
#문: 반환값X / 식: 반환값O
#대입 연산자 우측에 표현(식)이 와야 하는데 우측 ()안에 대입문을 상용해서 에러 발생
```

```python
#8-2. 월러스 연산자

n = 10
result = (x := n>5)   # 대입하면서 값 반환
print(result)
# := 월러스 연산자
```

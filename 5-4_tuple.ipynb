```python
#1. 튜플의 패킹/언패킹

t = 1, 3, 5 # 패킹 실행
print(t) 

l = list(t)
print(l, type(l))

x, y, z = t # 언패킹 실행
print(x, y, z)


a, b = [3, 6] #(3, 6) 변경 후 언패킹 발생
print(a, b)
```

```python
x = (1, 3)
print(x)

y = list(x)
print(y)

z = tuple(y)
print(z)
```

```python
# 패킹, 언패킹시 * 연산자
# a, b = [1, 2, 3, 4] #too many values to unpack

*a, b = [1, 2, 3, 4]
print(a)
print(b)

a, *b = (1, 2, 3, 4)
print(a)
print(b)

e, *f, g = (1, 2, 3, 4)
print(e)
print(f)
print(g)

h = [1, 2, 3]
print(h)
print(*h) #언패킹 == h[0], h[1], h[2]
```

```python
#조건표현식 (삼항연산자)

age = 20
result = "성인" if age >= 19 else "미성년자"
print(result)

age = 17
result = "성인" if age >= 19 else "미성년자"
print(result)

print(f"당신은 {'성인' if age >= 19 else '미성년자'} 입니다") # ""안에 문자열은 ''를 사용하자
```

```python
age = 15
result = ""
if age >= 19:
    result = "성인"
else:
    result = "미성년자"
print(result)
```

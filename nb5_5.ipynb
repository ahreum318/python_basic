```python
#set : 중복을 허용하지 않는 자료형, 순서가 없음 (인덱싱 불가)

a = {1, 2, 3, 2}
print(a) #중복값 제외하고 보여줌

print(list(a)[0]) #리스트 형태로 바꿔서 순서 만듦
print(a[0]) #순서 개념이 없어 에러 발생 ↑위와 같이 리스트 형태로 만들면 됨
```

```python
#1. 셋 집합연산

a = {1, 2, 3}
b = {3, 4, 5}

# 수학용 집합 연산이 가능하다
print(a & b) #교집합
print(a.intersection(b))

print(a | b) #합집합
print(a.union(b))

print(a - b) #차집합
print(a.difference(b))
```

```python
#2. 셋 내장함수
s = {1, 2, 3}

s.add(4) #추가
print(s)

s.update([5,6]) #여러 개 추가
print(s)

s.remove(2) # 값 2 삭제(없으면 오류)
print(s)

s.discard(10) # 값 없으면 무시


s = {1, 2, 3}
print(len(s)) #개수
print(2 in s) #포함 여부

s.clear() #전체 삭제
print(s) #set()
```

```python
fs = frozenset([1, 2, 3])
print(fs)
fs.add(4) #frozenset= 읽기 전용 set으로 만든 거여서 수정 안됨
```
